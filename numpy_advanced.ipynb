# In-Place-Operationen mit NumPy-ufuncs / *In-Place Operations using NumPy ufuncs*


In NumPy verändern In-Place-Operationen ein bestehendes Array, ohne eine neue Kopie zu erstellen, wodurch sie speichereffizienter und schneller sind.

Viele ufuncs unterstützen In-Place-Aktualisierungen über Parameter wie `out`.

--- 

*In NumPy, in-place operations modify an existing array without creating a new copy, making them more memory-efficient and faster.*

*Many ufuncs support in-place updates using parameters like `out`.*

````python
# Create a NumPy array
arr = np.array([2, 4, 6])
print(np.multiply(arr, 2))
print(arr)
# multiply each element by 2 and store in arr
np.multiply(arr, 2, out=arr)
print(arr)
````

````python
# Create a NumPy array
array_x = np.array([1, 2, 3, 4, 5])

# Create another array to add
array_y = np.array([10, 20, 30, 40, 50])

# Perform in-place addition using np.add with the out parameter
np.add(array_x, array_y, out=array_x)

# Print the resulting array
print("Resulting Array after In-Place Addition:")
print(array_x)
````

# Erstellen von eigenen ufunc / *How To Create Your Own ufunc*

Um eine eigene ufunc zu erstellen, müssen wir zunächst eine Funktion definieren, wie bei normalen Python-Funktionen. 
Anschließend fügen wir sie mit der Methode frompyfunc() zur NumPy-ufunc-Bibliothek hinzu.

Die Methode frompyfunc() nimmt die folgenden Argumente:
* `function` – der Name der Funktion
* `inputs` – die Anzahl der Eingabeargumente (Arrays)
* `outputs` – die Anzahl der Ausgabearrays
  
---

To create your own ufunc, you have to define a function, like you do with normal functions in Python, then you add it to your NumPy ufunc library with the `frompyfunc()` method.

The `frompyfunc()` method takes the following arguments:
- `function` - the name of the function.
- `inputs` - the number of input arguments (arrays).
- `outputs` - the number of output arrays.


````python
def custom_add(x, y):
  return x+y

custom_add = np.frompyfunc(custom_add, 2, 1)

print(type(custom_add)) 
````

````python
array_x = np.array([1, 2, 3, 4, 5])
array_y = np.array([10, 20, 30, 40, 50])

array_sum = custom_add(array_x, array_y)
print(array_sum)
````

### NumPy Linear Algebra Module

NumPy stellt ein spezielles Modul namens `np.linalg` bereit, um lineare Algebra-Operationen durchzuführen, wie z. B. das Lösen von Gleichungssystemen, das Berechnen von Determinanten, Inversen und Vektornormen. Es wird häufig in Data Science, Machine Learning und wissenschaftlichem Rechnen verwendet.

```
A = np.array([[1, 2],[3, 4]])
b = np.array([5, 6])
```

| Funktion                  | Beschreibung                        | Beispielcode               | Ausgabe                      |
| ------------------------- | ----------------------------------- | -------------------------- | ---------------------------- |
| `np.linalg.det()`         | Determinante einer Matrix           | `np.linalg.det(A)`         | `-2.0`                       |
| `np.linalg.inv()`         | Inverse einer Matrix                | `np.linalg.inv(A)`         | `[[-2. , 1. ], [1.5, -0.5]]` |
| `np.linalg.solve()`       | Lösen eines linearen Systems Ax = b | `np.linalg.solve(A, b)`    | `[ -4. , 4.5 ]`              |
| `np.linalg.eig()`         | Eigenwerte & Eigenvektoren          | `np.linalg.eig(A)`         | Werte + Vektoren             |
| `np.linalg.norm()`        | Vektor-/Matrixnorm                  | `np.linalg.norm([3,4])`    | `5.0`                        |
| `np.linalg.matrix_rank()` | Rang einer Matrix                   | `np.linalg.matrix_rank(A)` | `2`                          |
| `np.linalg.trace()`       | Summe der Diagonalelemente          | `np.trace(A)`              | `5`                          |
| `np.linalg.svd()`         | Singulärwertzerlegung               | `np.linalg.svd(A)`         | U, S, V Matrizen             |
| `np.linalg.pinv()`        | Pseudo-Inverse                      | `np.linalg.pinv(A)`        | Matrix                       |
| `np.linalg.cond()`        | Konditionszahl                      | `np.linalg.cond(A)`        | Zahlenwert                   |


---

NumPy provides a dedicated module called `np.linalg` for performing linear algebra operations such as solving equations, computing determinants, inverses, and vector norms. It is widely used in data science, machine learning, and scientific computing.

```

b = np.array([5, 6])
```

| `np.linalg.matrix_rank()` | Rank of matrix               | `np.linalg.matrix_rank(A)` | `2`                          |
| `np.linalg.trace()`       | Sum of diagonal elements     | `np.trace(A)`              | `5`                          |
| `np.linalg.svd()`         | Singular Value Decomposition | `np.linalg.svd(A)`         | U, S, V matrices             |
| `np.linalg.pinv()`        | Pseudo-inverse               | `np.linalg.pinv(A)`        | matrix                       |
